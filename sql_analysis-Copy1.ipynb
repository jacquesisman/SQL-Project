#### Primary Question
##### Is there a commonality between the content provided with the duration it has to how long it is being played by users

```python
%%sql
CREATE OR REPLACE VIEW report AS 
SELECT c.title, u.content_id, u.duration_played_total, c.duration, c.description, c.type,
CASE 
    WHEN c.duration < 500 THEN 'Short Length'
    WHEN c.duration <= 1000 THEN 'Medium Length'
    WHEN c.duration >= 2000 THEN 'Long Length'
END AS DurationReport
FROM ContentStartUp c
JOIN UserActivityStartUp u ON c.catalog_id = u.catalog_id
GROUP BY c.title;

SELECT * 
FROM report
WHERE type = 'original'
ORDER BY duration_played_total DESC
LIMIT 100;
```

#### Discovery 
In general the market seems to enjoy long durations of podcasts and other sources of media possibly. Having the duration played being ordered from the most you can clearly see that users like long durations in this specific market. This could be valid for many reasons for example traffic and being stuck around for a long time can lead to users choosing longer podcasts. And most of these podcasts are originals meaning what they are doing is working for the market. What they have been creating have been creative and taken well by the public.

#### Two supporting SQL for the primary question
1 - What is the genre type that these content most liked by users
2 - comparing witht the average duration are these content most liked above this average. 

```python
%%sql
WITH cte_startup_report (id, genre, content, event, entity, catalog, event_time) AS (
    SELECT c.id, c.genre_id, c.content_id, u.event_name, u.entity_subtype, u.catalog_id, u.event_timestamp
    FROM ContentGenreStartUp c
    JOIN UserActivityStartUp u ON c.content_id = u.content_id
)
SELECT 
    id, 
    genre,
    content,
    event,
    entity, 
    catalog,
    event_time
FROM cte_startup_report
LIMIT 50;
```

#### Discovery 
In this supporting SQL we can discovery what are the specific types of genres and content being most sought after in the market currently. This will allow a deeper dive into what the market wants in terms of creating topics or stories they want to learn/hear about. 

```python
%%sql
SELECT 
	title,
	id,
	catalog_id,
	status, 
	type,
	description,
	duration,
	(
		SELECT AVG(duration)
		FROM ContentStartUp
	) AS avg_duration,
	duration > 
	(
		SELECT AVG(duration)
		FROM ContentStartUp
	) AS greater_than_avg_duration
FROM ContentStartUp csu;
```

#### Discovery 
In this supporting SQL further supporting the first SQL using the average duration and calculating whether or not the duration average was greater than or less than it shows that it supports that mainly what is sought after in this market are long durationed podcasts. However with more data and specific data this can be further discovered again to the reasoning behind this driving force.
